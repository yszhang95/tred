```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.load("waveforms.npz")
```

```python
f.files
```

```python
c = np.squeeze(f['current_tpc0_batch0'])
pixc = f['current_tpc0_batch0_location'][:,:2]
tc = f['current_tpc0_batch0_location'][:,2]
```

```python
for i in range(c.shape[0]):
    if np.all(c[i]<1E-5):
        continue
    t = tc[i] + np.arange(c[i].shape[-1])
    plt.plot(t, c[i], label=f'PixInd ({pixc[i,0]}, {pixc[i,1]})')
plt.xlim(-100,200)
plt.xlabel('time ticks [/50ns]')
plt.ylabel('charge per 50ns [ke-]')
plt.title("induced current waveform")
plt.legend()
```

```python
for i in range(c.shape[0]):
    if np.all(c[i]<1E-5):
        continue
    pm = np.all(f['effq_tpc0_batch0_location'][:,:2] == pixc[i], axis=1)
    plt.plot(f['effq_tpc0_batch0_location'][:,-1][pm], f['effq_tpc0_batch0'][pm][:,-1]/1E3, label=f'PixInd ({pixc[i,0]}, {pixc[i,1]})')
plt.xlim(-100,200)
plt.xlabel('time ticks to arrive at anode [/50ns]')
plt.ylabel('charge per 50ns [ke-]')
plt.title("ionization charge")
plt.legend()
```
