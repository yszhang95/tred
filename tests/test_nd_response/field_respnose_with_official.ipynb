```python
import os
def load_data(directory):
    response = np.load(os.path.join(directory, "fr_4p4pitch_3.8pix_nogrid_10pathsperpixel.npy"))
    start = np.load(os.path.join(directory, "startpoints.npy"))
    end = np.load(os.path.join(directory, "endpoints.npy"))
    return response, start, end
```

```python
response, start, end = load_data(".")
```

```python
pts = np.vstack([start[:,0], start[:,1]]).T
P = np.asarray(pts)           # shape (N, 2)

# starts (all but last) and deltas to next point
x0, y0 = P[:-1, 0], P[:-1, 1]
dx, dy = np.diff(P[:, 0]), np.diff(P[:, 1])

plt.figure(figsize=(8, 8))
q = plt.quiver(x0, y0, dx, dy,
               angles="xy", scale_units="xy", scale=1,   # lengths in data units
               width=0.003, headwidth=4, headlength=6)
plt.plot(P[:, 0], P[:, 1], linestyle="--", alpha=0.5)    # optional: path
plt.scatter(P[:, 0], P[:, 1], s=25)                      # optional: points

plt.axis("equal")
plt.xlabel("x"); plt.ylabel("y")
plt.title("Arrows from (x[i], y[i]) -> (x[i+1], y[i+1])")
#plt.xlim(0, 24)
# plt.ylim(0, 24)
plt.tight_layout()
plt.show()
print(np.min(start[:,0]), np.max(start[:,0]))
print(np.min(start[:,0]), np.max(start[:,0]))
```
