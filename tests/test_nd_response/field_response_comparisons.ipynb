# Comparisons between FR in TRED paper and my reproduction
## Geometry in the two setup are mismatching
### Geometry in FR in TRED paper
Field response is calculated at a `5x5` pixel array. The field response is the top-right quarter of the pixel array. The collection pixel is only saved in a quarter, for indices [0, 4), [25, 30), and so on. One can get the collection by `fr.reshape(5, 5, 5, 5, -1)[0, :, 0, :, :]`
### Geoemtry in my reproduction.
Field response is calculated at a `9x9` pixel array. The field response is the lower-left quarter of the pixel array. The collection pixel is fully saved, for indices [615, 625), [575, 600), and so on. The counterpart for the top-right quarter in TRED is `fr.reshape(5, 5, 5, 5, -1)[-1, :, -1, :, :]`. The counterpart of the +1 adjacent pixel `fr_tred.reshape(5, 5, 5, 5, -1)[1, :, 0, :, :]` in TRED is `fr.reshape(5, 5, 5, 5, -1)[-3, ::-1, -1, ::-1, :]`.
